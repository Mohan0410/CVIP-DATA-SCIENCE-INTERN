```python
data['Glucose']=data['Glucose'].replace(0,data['Glucose'].mean())
data['BloodPressure']=data['BloodPressure'].replace(0,data['BloodPressure'].mean())
data['SkinThickness']=data['SkinThickness'].replace(0,data['SkinThickness'].mean())
data['Insulin']=data['Insulin'].replace(0,data['Insulin'].mean())
data['BMI']=data['BMI'].replace(0,data['BMI'].mean())
```

```python
x=data.drop('Outcome',axis=1)
y=data['Outcome']
```

```python
x
```

```python
y
```

```python
scaler = StandardScaler()
scaler.fit(x)
standardized_data = scaler.transform(x)
x = standardized_data
y = data['Outcome']
```

```python

xtr,xte,ytr,yte=tts(x,y,test_size=0.20,random_state=42)
```

```python
p_lr=Pipeline([('scalar1',StandardScaler()),('lr_classifier',LogisticRegression())])
p_knn=Pipeline([('scalar2',StandardScaler()),('knn_classifier',KNeighborsClassifier())])
p_svc=Pipeline([('scalar3',StandardScaler()),('svc_classifier',SVC())])
p_dt=Pipeline([('dt_classifier',DecisionTreeClassifier())])
p_rf=Pipeline([('rf_classifier',RandomForestClassifier(max_depth=3))])
p_gbc=Pipeline([('gbc_classifier',GradientBoostingClassifier())])
```

```python
pl=[p_lr,p_knn,p_svc,p_dt,p_rf,p_gbc]
```

```python
pl
```

```python
for pipe in pl:
    pipe.fit(xtr,ytr)
```

```python
pd={0:'LR',
    1:'KNN',
    2:'SVC',
    3:'DT',
    4:'RF',
    5:'GBC'
   }
```

```python
pd
```

```python
for i,model in enumerate(pl):
    print("{} test acc:{}".format(pd[i],model.score(xte,yte)))
```
